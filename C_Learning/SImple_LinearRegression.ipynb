## **Linear Regression**

### **Evaluate Model performance**

The provided code calculates the mean squared error (MSE) and R-squared (R2) scores to evaluate the performance of the linear regression model on both the training and testing datasets. Mean squared error quantifies the average squared difference between the predicted and actual target variable values, providing a measure of prediction accuracy. R-squared represents the proportion of the variance in the target variable that is explained by the model, indicating its overall goodness of fit. The calculated metrics, stored in variables like `lr_train_mse`, `lr_train_r2`, `lr_test_mse`, and `lr_test_r2`, offer insights into the model's accuracy and explanatory power on both the training and testing datasets. Lower MSE values and higher R2 scores are indicative of better model performance.

```python
from sklearn.metrics import mean_squared_error , r2_score

lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)
```

The code creates a Pandas DataFrame, `lr_results`, to store and present the performance metrics of the linear regression model.
```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Traning R2', 'Testing MSE', 'Testing R2']
lr_results
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import RandomizedSearchCV

lasso_params = {
    'alpha': [0.001, 0.01, 0.1, 1.0, 10.0],  # Regularization strength
}

lasso_random_search = RandomizedSearchCV(Lasso(random_state=42), lasso_params, cv=5, n_iter=1000, scoring='neg_mean_squared_error', random_state=42)
lasso_random_search.fit(X_train, y_train)

best_lasso_params = lasso_random_search.best_params_
print("Best Hyperparameters for Lasso Regression:", best_lasso_params)

# Train the final Lasso Regression model with the best hyperparameters
final_lasso_model = Lasso(**best_lasso_params, random_state=42)
final_lasso_model.fit(X_train, y_train)
```

The provided code performs a randomized search for hyperparameter tuning of a random forest regression model using scikit-learn's `RandomizedSearchCV`. It explores different combinations of hyperparameters specified in the `rf_params` dictionary, including the number of estimators, maximum depth, minimum samples split, and minimum samples leaf. The search is conducted using 5-fold cross-validation (`cv=5`) and aims to find the hyperparameter combination that minimizes the negative mean squared error (`scoring='neg_mean_squared_error'`). The `n_iter=10` parameter controls the number of random combinations to try. After fitting the model with the training data (`X_train` and `y_train`), the best hyperparameters are printed, providing guidance for optimizing the model's performance.

```python
from sklearn.model_selection import RandomizedSearchCV

rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_random_search = RandomizedSearchCV(RandomForestRegressor(random_state=42), rf_params, cv=5, n_iter=10, scoring='neg_mean_squared_error', random_state=42)
rf_random_search.fit(X_train, y_train)

best_rf_params = rf_random_search.best_params_
print("Best Hyperparameters for Random Forest:", best_rf_params)
```

## **Feature Selection**

The provided code conducts feature selection using Recursive Feature Elimination (RFE) with two regression models: Linear Regression and Random Forest. For Linear Regression, it initializes RFE with an estimator and fits it to the training data, then identifies and prints the selected features. The final Linear Regression model is trained using these selected features. Similarly, for Random Forest, RFE is initialized with a RandomForestRegressor using optimal hyperparameters, and the selected features are identified. The final Random Forest model is then trained using the chosen features. This process aids in selecting the most relevant features for each model, potentially enhancing their predictive performance.

```python
import numpy as np
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Assuming X_train is a NumPy array, and feature_names is a list of your column names
feature_names = ['Number_of_Customers','Menu_Price','Marketing_Spend','Average_Customer_Spending','Promotions','Reviews']

# Linear Regression
rfe_lr = RFE(estimator=LinearRegression(), n_features_to_select=4)
rfe_lr.fit(X_train, y_train)

selected_features_lr = np.array(feature_names)[rfe_lr.support_]
print("Selected Features for Linear Regression:", selected_features_lr)

final_lr_model = LinearRegression()
final_lr_model.fit(X_train[:, rfe_lr.support_], y_train)

# Random Forest
rfe_rf = RFE(estimator=RandomForestRegressor(**best_rf_params, random_state=42), n_features_to_select=4)
rfe_rf.fit(X_train, y_train)

selected_features_rf = np.array(feature_names)[rfe_rf.support_]
print("Selected Features for Random Forest:", selected_features_rf)

final_rf_model = RandomForestRegressor(**best_rf_params, random_state=42)
final_rf_model.fit(X_train[:, rfe_rf.support_], y_train)
```

The provided code evaluates the performance of the final Linear Regression and Random Forest models on the testing dataset after feature selection using Recursive Feature Elimination (RFE). For Linear Regression, it predicts the target variable (`y_pred_lr_final`) using the selected features and calculates the Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared (R2) score. The process is repeated for Random Forest, with predictions (`y_pred_rf_final`) and corresponding evaluation metrics calculated. These metrics offer insights into the models' accuracy and goodness of fit on the unseen data, providing a comprehensive assessment of their performance after the feature selection process.

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predictions for Linear Regression with selected features
y_pred_lr_final = final_lr_model.predict(X_test[:, rfe_lr.support_])

# Mean Squared Error for Linear Regression after RFE
mse_lr_final = mean_squared_error(y_test, y_pred_lr_final)
print("Mean Squared Error (Linear Regression - Final):", mse_lr_final)

# Root Mean Squared Error for Linear Regression after RFE
rmse_lr_final = np.sqrt(mse_lr_final)
print("Root Mean Squared Error (Linear Regression - Final):", rmse_lr_final)

# R2 Score for Linear Regression after RFE
lr_final_r2 = r2_score(y_test, y_pred_lr_final)
print("R2 Score (Linear Regression - Final):", lr_final_r2)

print("\n\n")

# Predictions for Random Forest with selected features
y_pred_rf_final = final_rf_model.predict(X_test[:, rfe_rf.support_])

# Mean Squared Error for Random Forest after RFE
mse_rf_final = mean_squared_error(y_test, y_pred_rf_final)
print("Mean Squared Error (Random Forest - Final):", mse_rf_final)

# Root Mean Squared Error for Random Forest after RFE
rmse_rf_final = np.sqrt(mse_rf_final)
print("Root Mean Squared Error (Random Forest - Final):", rmse_rf_final)

# R2 Score for Random Forest after RFE
rf_final_r2 = r2_score(y_test, y_pred_rf_final)
print("R2 Score (Random Forest - Final):", rf_final_r2)
```

The provided code makes a comparison between the Root Mean Squared Error (RMSE) of the final Linear Regression model (`rmse_lr_final`) and the final Random Forest model (`rmse_rf_final`) after feature selection. If the RMSE of Linear Regression is less than that of Random Forest, it prints "Linear regression performs better after feature selection." Otherwise, it prints "Random Forest performs better after feature selection." This conditional statement helps determine which model exhibits better performance in terms of prediction accuracy on the testing dataset after the feature selection process.
