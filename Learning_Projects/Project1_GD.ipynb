```python
df = df0[['Alcohol', 'HepatitisB', 'Measles', 'BMI', 'Polio', 'HIV/AIDS', 'Lifeexpectancy']]

# Display the new DataFrame
df.head()
```

### Checking for missing values presents in Columns

```python
total_missing_values = 0
for col in df.columns:
    missing_values = df[col].isnull().sum()
    total_missing_values += missing_values
    print(f'Missing values on {col} is {missing_values}.')
print(f'Total Missing values {total_missing_values}.')
```

### Filling the missing or null values with the values of previous column using method <span style="color:green">ffill<span>

```python
total_missing_values = 0
for col in df.columns:
    missing_values = df[col].isnull().sum()
    if missing_values > 0:
        df[col].fillna(method='ffill', inplace=True)  # Use inplace=True to modify the DataFrame in place
    total_missing_values += missing_values
    print(f'Missing values in {col} is {missing_values}.')
print(f'Total Missing values: {total_missing_values}.')
```

# Multiple Linear Regression


## Using <span style='color:green'>Sklearn</span> <span style='color:yellow'>Ordinary Least Squares (OLS) </span>

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X = df.iloc[: , :-1]
y = df.iloc[: , 1]
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
df.head()
```

```python
y_pred = lr.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

## Using <span style='color:green'>Sklearn</span> <span style='color:yellow'>Stochastic Gradient Descent </span>
