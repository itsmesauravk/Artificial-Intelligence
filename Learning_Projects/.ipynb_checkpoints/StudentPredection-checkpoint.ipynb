```python
# with normalization

import numpy as np

class GDRegression:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.coef_ = None
        self.intercept_ = None
        self.epochs = epochs
        self.l_rate = learning_rate

    def fit(self, X, y):
        # Normalize the features
        X_mean = np.mean(X, axis=0)
        X_std = np.std(X, axis=0)
        X = (X - X_mean) / X_std

        # Initialize the intercept and coefficients
        self.intercept_ = 0  
        self.coef_ = np.zeros(X.shape[1])

        for i in range(self.epochs):
            # Vectorization for y prediction
            y_hat =  np.dot(X, self.coef_) + self.intercept_  # y_hat = mx + b
            
            # Calculating intercept derivative
            intercept_der = -2 * np.mean(y - y_hat)
            self.intercept_ = self.intercept_ - (self.l_rate * intercept_der)
            
            # Calculating coefficient derivatives
            coef_der = -2 * np.dot((y - y_hat), X) / X.shape[0]
            self.coef_ = self.coef_ - (self.l_rate * coef_der)

    def predict(self, X):
        # Normalize the features using the same mean and std as training data
        X_mean = np.mean(X, axis=0)
        X_std = np.std(X, axis=0)
        X = (X - X_mean) / X_std
        
        return np.dot(X, self.coef_) + self.intercept_

```

```python
#making instance 
gdr = GDRegression(learning_rate=0.01, epochs=1000)
```

```python
#training model
gdr.fit(X_train,y_train)
```

```python
y_pred_gdr = gdr.predict(X_test)
```

```python
r2_gdr = r2_score(y_test,y_pred_gdr)
r2_gdr
```

```python
print("Linear Regression r2 score : ", r2_lr)
print("Gradient Descent r2 score : ", r2_gdr)
```

## Using <span style='color:green'>Sklearn </span>  With <span style='color:yellow'>Stochastic Gradient Descent </span>

```python
from sklearn.linear_model import SGDRegressor
```

```python
# creating instance
sgdr = SGDRegressor(max_iter=100, 
                    learning_rate='optimal', 
                    eta0=0.1,   
                    alpha=0.01, 
                    tol=1e-3, 
                    random_state=42)  #eta0 -> initial learning rate, tol -> tolerence (means stop the loop if prev and new diff is less then given)
```

```python
sgdr.fit(X_train_n,y_train)
```

```python
y_pred_sgdr = sgdr.predict(X_test_n)
```

```python
r2_sgdr = r2_score(y_test,y_pred_sgdr)
r2_sgdr
```
