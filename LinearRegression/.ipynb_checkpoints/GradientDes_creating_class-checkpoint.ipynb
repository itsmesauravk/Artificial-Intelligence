```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples= 100, n_features= 1, n_targets = 1, n_informative = 1 , noise=30)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state = 2)
```

```python
plt.scatter(X_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)

print("Value of m : ",lr.coef_)
print("Value of b : ",lr.intercept_)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

### Creating Class 


```python
class GDRegression:
    def __init__(self,learning_rate,epochs):
        self.m = -100
        self.b = 60
        self.l_r = learning_rate
        self.ep = epochs

    def fit(self, X,y):
        #calculating b using Gradient descent
        for i in range(self.ep):
            loss_slope_b = -2 * np.sum(y - self.m * X.ravel() - self.b)
            loss_slope_m = -2 * np.sum((y - self.m * X.ravel() - self.b) * X.ravel())
            
            self.b = self.b - (self.l_r * loss_slope_b)
            self.m = self.m - (self.l_r * loss_slope_m)
            
        print("b :",self.b,"m :", self.m)
        

    def predict(self,X):
        y_pred = (self.m * X) + self.b
        return y_pred
        
```

```python
l_r = 0.001
epochs = 50
```

```python
Glr = GDRegression(l_r, epochs)
```

```python
Glr.fit(X_train,y_train)
```

```python
y_pred_gd = Glr.predict(X_test)
```

```python
r2_score(y_test, y_pred_gd)
```
